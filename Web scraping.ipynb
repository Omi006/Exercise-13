```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
# URL of the page to scrape
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'  # Replace with the actual URL
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
revenue_table = soup.find('table')  
rows = revenue_table.find_all('tr')
revenue_data = []
for row in rows[1:]:  
    cols = row.find_all('td')
    revenue_data.append([col.text.strip() for col in cols])
tesla_revenue = pd.DataFrame(revenue_data, columns=['Year', 'Revenue'])
print(tesla_revenue.tail())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch Tesla stock data
tesla_data = yf.download("TSLA", start="2020-01-01", end="2023-09-28")

# Reset index if necessary
tesla_data.reset_index(inplace=True)

def make_graph(dataframe):
    plt.figure(figsize=(10, 5))  # Set the figure size
    plt.plot(dataframe['Date'], dataframe['Close'], label='Tesla Stock Price', color='blue')
    plt.title('Tesla Stock Price Over Time')  # Title of the graph
    plt.xlabel('Date')  # X-axis label
    plt.ylabel('Stock Price (USD)')  # Y-axis label
    plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
    plt.legend()  # Show legend
    plt.tight_layout()  # Adjust layout
    plt.show()  # Show the plot

# Call the function with the tesla_data DataFrame
make_graph(tesla_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

# Fetch GameStop stock data
gme_data = yf.download("GME", start="2020-01-01", end="2023-09-28")

# Reset index if necessary
gme_data.reset_index(inplace=True)

def make_graph(dataframe, title):
    plt.figure(figsize=(10, 5))  # Set the figure size
    plt.plot(dataframe['Date'], dataframe['Close'], label='GameStop Stock Price', color='orange')
    plt.title(title)  # Title of the graph
    plt.xlabel('Date')  # X-axis label
    plt.ylabel('Stock Price (USD)')  # Y-axis label
    plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
    plt.legend()  # Show legend
    plt.tight_layout()  # Adjust layout
    plt.show()  # Show the plot

# Call the function with the gme_data DataFrame
make_graph(gme_data, 'GameStop Stock Price Over Time')
```
